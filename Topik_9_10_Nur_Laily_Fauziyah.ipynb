## Introduction to Probability and Statistics
## Assignment

In this assignment, we will use the dataset of diabetes patients taken [from here](https://www4.stat.ncsu.edu/~boos/var.select/diabetes.html).

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('https://www4.stat.ncsu.edu/~boos/var.select/diabetes.tab.txt', sep="\t")
df.head()
```


In this dataset, columns as the following:
* Age and sex are self-explanatory
* BMI is body mass index
* BP is average blood pressure
* S1 through S6 are different blood measurements
* Y is the qualitative measure of disease progression over one year

Let's study this dataset using methods of probability and statistics.

### Task 1: Compute mean values and variance for all values

```python
# MEAN
df[['AGE','SEX','BMI','BP', 'S1','S1','S2','S3','S4','S5','S6','Y']].mean()
```

```python
# VARIANCE
df[['AGE','SEX','BMI','BP', 'S1','S1','S2','S3','S4','S5','S6','Y']].var()
```

### Task 2: Plot boxplots for BMI, BP and Y depending on gender

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))

#Membuat sub boxplot masing-masing varuabel
sns.boxplot(x='SEX', y='BMI', data=df, ax=axes[0, 0])
axes[0, 0].set_title('BMI')

sns.boxplot(x='SEX', y='BP', data=df, ax=axes[0, 1])
axes[0, 1].set_title('BP')

sns.boxplot(x='SEX', y='Y', data=df, ax=axes[1, 0])
axes[1, 0].set_title('Y')

# Menghilangkan subplot yang tidak terpakai
fig.delaxes(axes[1, 1])

# Menampilkan subplot
plt.tight_layout()
plt.show()
```

### Task 3: What is the the distribution of Age, Sex, BMI and Y variables?

```python
#Menampilkan distribusi menggunakan histogram, dengan langkah-langkah :
#1 Membuat grid subplot
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))

#2 Membuat sub histogram masing-masing variabel
sns.histplot(data=df, x='AGE', ax=axes[0, 0], kde=True)
axes[0, 0].set_title('AGE')

sns.histplot(data=df, x='SEX', ax=axes[0, 1], kde=True)
axes[0, 1].set_title('SEX')

sns.histplot(data=df, x='BMI', ax=axes[1, 0], kde=True)
axes[1, 0].set_title('BMI')

sns.histplot(data=df, x='Y', ax=axes[1, 1], kde=True)
axes[1, 0].set_title('Y')


#3 Menampilkan subplot
plt.tight_layout()
plt.show()
```
