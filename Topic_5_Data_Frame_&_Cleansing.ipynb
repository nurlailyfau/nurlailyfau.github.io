```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset('titanic')
df.head()
```

```python
df = df.dropna(subset=['embark_town']) #drop baris
df.isnull().sum()
```

```python
df = df.drop('deck', axis=1) #drop kolom -> kalo banyak banget yang error
df.isnull().sum()
```

```python
#2 Kalo datanya numerik, Mengisi data dengan mean, median, min, max, 0
```

```python
df[['age','fare']].describe() #jika distribusi normal (mean & med selisih tidak banyak), gunakan median/mean
```

```python
sns.distplot(df['age']) #-> distribusi normal
```

```python
sns.distplot(df['fare']) #-> distribusi tidak normal, gunakan MEDIAN
```

```python
df['age'] = df['age'].fillna(df['age'].mean())
df.isnull().sum()
```

```python
#3 Kalau data kategori, isi dengan unknown atau yang terbanyak
```

```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset('titanic')
df.head()
```

```python
df.isnull().sum()
```

```python
# Mengisi dengan unknown atau others jika datanya sangat banyak
```

```python
df['deck'] = np.where(df['deck'].isnull(),"Unk",df['deck'])
df.isnull().sum()
df.head()
```

```python
# Mengisi dengan modus jika datanya sedikit
```

```python
df['embark_town'].value_counts()
```

```python
df['embark_town'] = df['embark_town'].fillna(df['embark_town'].mode())
df.head()
```

```python
# DATA OUTLIER
```

```python
sns.boxplot(df['fare'],color="green",orient='h')
```

```python
#Solusi
#1 Logaritma
```

```python
df['log_fare'] = np.log(df['fare']+1)
```

```python
#2 Z score
```

```python
from scipy import stats 
```

```python
df['fare_zscore'] = abs(stats.zscore(df['fare']))
df.head()
```

```python
df = df[df['fare_zscore'] < 3]
```

```python
sns.boxplot(df['fare'],color="green",orient='h')
```

```python
#3 IQR
```

```python
print('jumlah baris sebelum outlier pada Fare:', len(df))
Q1 = df['fare'].quantile(0.25)
Q3 = df['fare'].quantile(0.75)
IQR = Q3 - Q1
low_limit = Q1 - (1.5 * IQR)
high_limit = Q3 - (1.5 * IQR)
filtered_entries = ((df['fare'] >= low_limit) & (df['fare'] <= high_limit))
df = df[filtered_entries]
print(f'jumlah baris setelah filter outlier:', len(df))
df.head(5)
```
